## Clean Column Names and Handle Missing Values

```python
# Clean column names
print("Cleaning column names...")
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^a-zA-Z0-9_]', '', regex=True)

print(f"Target column after cleaning: {target_column.lower()}")
target_column = target_column.lower()

# Identify numeric and categorical columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Remove target from feature lists
if target_column in numeric_cols:
    numeric_cols.remove(target_column)
if target_column in categorical_cols:
    categorical_cols.remove(target_column)

print(f"\nNumeric columns: {len(numeric_cols)}")
print(f"Categorical columns: {len(categorical_cols)}")

# Handle infinite values
print("\nReplacing infinite values...")
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Remove columns with too many missing values (>50%)
threshold = 0.5
cols_to_drop = [col for col in df.columns if df[col].isnull().sum() / len(df) > threshold]
if cols_to_drop:
    print(f"\nDropping columns with >50% missing values: {cols_to_drop}")
    df.drop(cols_to_drop, axis=1, inplace=True)
    numeric_cols = [col for col in numeric_cols if col not in cols_to_drop]
    categorical_cols = [col for col in categorical_cols if col not in cols_to_drop]

# Fill missing values
print("\nHandling missing values...")
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0] if len(df[col].mode()) > 0 else 'unknown', inplace=True)

# Final check
df.fillna(0, inplace=True)

print(f"\nFinal dataset shape: {df.shape}")
print(f"Missing values after cleaning: {df.isnull().sum().sum()}")

# Memory optimization
print("\nOptimizing memory usage...")
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], downcast='float')

print(f"Memory after optimization: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

```python
# Separate features and target
X = df.drop(target_column, axis=1)
y = df[target_column]

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")

# Remove low variance features
print("\nRemoving low variance features...")
variance_threshold = VarianceThreshold(threshold=0.01)
X_high_var = variance_threshold.fit_transform(X)
selected_features = X.columns[variance_threshold.get_support()].tolist()

print(f"Features before: {X.shape[1]}")
print(f"Features after variance threshold: {len(selected_features)}")
print(f"Removed {X.shape[1] - len(selected_features)} features")

X = X[selected_features]

# Remove highly correlated features
print("\nRemoving highly correlated features (>0.95)...")
corr_matrix = X.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

if to_drop:
    print(f"Dropping {len(to_drop)} highly correlated features")
    X.drop(to_drop, axis=1, inplace=True)

print(f"\nFinal feature count: {X.shape[1]}")

# Save feature names
feature_names = X.columns.tolist()
with open('models/feature_names.pkl', 'wb') as f:
    pickle.dump(feature_names, f)

print(f"\nFeature names saved: {len(feature_names)} features")
```

## Cell 11: Advanced Class Balancing Methods (Complete Replacement)

```python
# Advanced Class Imbalance Handling
print("="*80)
print("ADVANCED CLASS IMBALANCE HANDLING")
print("="*80)

print(f"\nOriginal distribution:")
class_dist = pd.Series(y_train).value_counts().sort_index()
for cls, count in class_dist.items():
    print(f"  Class {cls}: {count:,} ({count/len(y_train)*100:.2f}%)")

imbalance_ratio = class_dist.max() / class_dist.min()
print(f"\nImbalance Ratio: {imbalance_ratio:.2f}:1")

# Get class information
minority_class = class_dist.idxmin()
majority_class = class_dist.idxmax()
minority_count = class_dist.min()
majority_count = class_dist.max()

print(f"\nMinority class: {minority_class} ({minority_count:,} samples)")
print(f"Majority class: {majority_class} ({majority_count:,} samples)")

# ============================================================================
# METHOD 1: Hybrid Sampling (Undersample + Duplicate with Noise)
# ============================================================================
print("\n" + "="*80)
print("METHOD 1: HYBRID SAMPLING (Undersample + Augmentation)")
print("="*80)

# Step 1: Undersample majority to reasonable ratio (e.g., 3:1)
target_ratio = 3  # Majority:Minority ratio
target_majority = minority_count * target_ratio

print(f"\nStep 1: Undersampling majority class to {target_majority:,} samples...")

minority_indices = np.where(y_train.values == minority_class)[0]
majority_indices = np.where(y_train.values == majority_class)[0]

# Random undersample majority
np.random.seed(RANDOM_STATE)
majority_sampled = np.random.choice(majority_indices, size=target_majority, replace=False)

# Step 2: Oversample minority with noise augmentation
print(f"\nStep 2: Augmenting minority class with noise...")

minority_oversample_count = target_majority  # Match majority count
repetitions = minority_oversample_count // minority_count
remainder = minority_oversample_count % minority_count

print(f"  Creating {minority_oversample_count:,} minority samples...")
print(f"  Repetitions: {repetitions}, Remainder: {remainder}")

# Get minority samples
X_minority = X_train_scaled[minority_indices]
y_minority = y_train.values[minority_indices]

# Create augmented minority samples with small Gaussian noise
X_minority_augmented = []
y_minority_augmented = []

for i in range(repetitions):
    # Add small random noise (0.5% of std)
    noise = np.random.normal(0, 0.005, X_minority.shape)
    X_augmented = X_minority + noise
    X_minority_augmented.append(X_augmented)
    y_minority_augmented.append(y_minority)

# Add remainder
if remainder > 0:
    indices_remainder = np.random.choice(minority_count, size=remainder, replace=False)
    noise = np.random.normal(0, 0.005, (remainder, X_minority.shape[1]))
    X_augmented = X_minority[indices_remainder] + noise
    X_minority_augmented.append(X_augmented)
    y_minority_augmented.append(y_minority[indices_remainder])

# Combine all minority samples
X_minority_final = np.vstack(X_minority_augmented)
y_minority_final = np.concatenate(y_minority_augmented)

# Combine majority and minority
X_majority_sampled = X_train_scaled[majority_sampled]
y_majority_sampled = y_train.values[majority_sampled]

X_train_balanced = np.vstack([X_majority_sampled, X_minority_final])
y_train_balanced = np.concatenate([y_majority_sampled, y_minority_final])

# Shuffle
shuffle_indices = np.random.permutation(len(X_train_balanced))
X_train_balanced = X_train_balanced[shuffle_indices]
y_train_balanced = y_train_balanced[shuffle_indices]

print(f"\n✓ Hybrid sampling completed!")
print(f"  Final shape: {X_train_balanced.shape}")
print(f"  Distribution:")
for cls, count in pd.Series(y_train_balanced).value_counts().sort_index().items():
    print(f"    Class {cls}: {count:,} ({count/len(y_train_balanced)*100:.2f}%)")

# ============================================================================
# Prepare MinMax scaled version
# ============================================================================
print("\n" + "="*80)
print("PREPARING MINMAX SCALED VERSION")
print("="*80)

# Apply same sampling to MinMax data
X_minority_minmax = X_train_minmax[minority_indices]

# Create augmented minority samples
X_minority_minmax_augmented = []

for i in range(repetitions):
    noise = np.random.normal(0, 0.005, X_minority_minmax.shape)
    X_augmented = X_minority_minmax + noise
    # Clip to [0, 1] range for MinMax
    X_augmented = np.clip(X_augmented, 0, 1)
    X_minority_minmax_augmented.append(X_augmented)

if remainder > 0:
    noise = np.random.normal(0, 0.005, (remainder, X_minority_minmax.shape[1]))
    X_augmented = X_minority_minmax[indices_remainder] + noise
    X_augmented = np.clip(X_augmented, 0, 1)
    X_minority_minmax_augmented.append(X_augmented)

X_minority_minmax_final = np.vstack(X_minority_minmax_augmented)

X_majority_minmax_sampled = X_train_minmax[majority_sampled]

X_train_minmax_balanced = np.vstack([X_majority_minmax_sampled, X_minority_minmax_final])
X_train_minmax_balanced = X_train_minmax_balanced[shuffle_indices]

print(f"✓ MinMax balanced data prepared: {X_train_minmax_balanced.shape}")

# ============================================================================
# Visualization
# ============================================================================
print("\n" + "="*80)
print("VISUALIZATION")
print("="*80)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Before balancing
counts_before = pd.Series(y_train).value_counts().sort_index()
axes[0].bar(range(len(counts_before)), counts_before.values, color='#FF6B6B', alpha=0.8, edgecolor='black')
axes[0].set_xlabel('Class', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Count (log scale)', fontsize=13, fontweight='bold')
axes[0].set_yscale('log')
axes[0].set_title('Before Balancing\n(Extreme Imbalance)', fontsize=14, fontweight='bold')
axes[0].set_xticks(range(len(counts_before)))
axes[0].grid(True, alpha=0.3, axis='y', linestyle='--')
for i, v in enumerate(counts_before.values):
    axes[0].text(i, v * 1.1, f'{v:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Imbalance ratio
axes[1].text(0.5, 0.6, f'Imbalance Ratio', ha='center', va='center', 
            transform=axes[1].transAxes, fontsize=16, fontweight='bold')
axes[1].text(0.5, 0.4, f'{imbalance_ratio:.1f} : 1', ha='center', va='center',
            transform=axes[1].transAxes, fontsize=32, fontweight='bold', color='red')
axes[1].text(0.5, 0.25, f'Majority: {majority_count:,}', ha='center', va='center',
            transform=axes[1].transAxes, fontsize=11, style='italic')
axes[1].text(0.5, 0.15, f'Minority: {minority_count:,}', ha='center', va='center',
            transform=axes[1].transAxes, fontsize=11, style='italic')
axes[1].axis('off')

# After balancing
counts_after = pd.Series(y_train_balanced).value_counts().sort_index()
axes[2].bar(range(len(counts_after)), counts_after.values, color='#4ECDC4', alpha=0.8, edgecolor='black')
axes[2].set_xlabel('Class', fontsize=13, fontweight='bold')
axes[2].set_ylabel('Count', fontsize=13, fontweight='bold')
axes[2].set_title('After Hybrid Balancing\n(Balanced)', fontsize=14, fontweight='bold')
axes[2].set_xticks(range(len(counts_after)))
axes[2].grid(True, alpha=0.3, axis='y', linestyle='--')
for i, v in enumerate(counts_after.values):
    axes[2].text(i, v * 1.02, f'{v:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('plots/03_advanced_balancing.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# Summary Statistics
# ============================================================================
print("\n" + "="*80)
print("SUMMARY")
print("="*80)

summary_data = {
    'Metric': ['Total Samples', 'Class 0 (Minority)', 'Class 1 (Majority)', 'Ratio (Maj:Min)'],
    'Before': [
        f"{len(y_train):,}",
        f"{counts_before[minority_class]:,} ({counts_before[minority_class]/len(y_train)*100:.2f}%)",
        f"{counts_before[majority_class]:,} ({counts_before[majority_class]/len(y_train)*100:.2f}%)",
        f"{imbalance_ratio:.2f}:1"
    ],
    'After': [
        f"{len(y_train_balanced):,}",
        f"{counts_after[minority_class]:,} ({counts_after[minority_class]/len(y_train_balanced)*100:.2f}%)",
        f"{counts_after[majority_class]:,} ({counts_after[majority_class]/len(y_train_balanced)*100:.2f}%)",
        f"{counts_after[majority_class]/counts_after[minority_class]:.2f}:1"
    ]
}

summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False))

# Save balanced data info
with open('results/balancing_info.txt', 'w') as f:
    f.write("="*80 + "\n")
    f.write("CLASS BALANCING INFORMATION\n")
    f.write("="*80 + "\n\n")
    f.write(f"Method: Hybrid Sampling (Undersample + Noise Augmentation)\n")
    f.write(f"Target Ratio: {target_ratio}:1\n")
    f.write(f"Noise Level: 0.5% of std\n\n")
    f.write(summary_df.to_string(index=False))
    f.write("\n\n")
    f.write("="*80 + "\n")

print("\n✓ Balancing information saved to 'results/balancing_info.txt'")

# Memory cleanup
gc.collect()

print("\n" + "="*80)
print("✓ ADVANCED CLASS BALANCING COMPLETED!")
print("="*80)
```

## Attention Mechanism Implementation

```python
# Attention mechanisms for LSTM and GRU
print("="*80)
print("DEFINING ATTENTION MECHANISMS")
print("="*80)

class AttentionLayer(nn.Module):
    """
    Simple Attention mechanism
    """
    def __init__(self, hidden_dim):
        super(AttentionLayer, self).__init__()
        self.hidden_dim = hidden_dim
        self. attention = nn.Linear(hidden_dim, 1)
        
    def forward(self, lstm_output):
        # lstm_output shape: (batch_size, seq_len, hidden_dim)
        
        # Calculate attention weights
        attention_weights = torch.softmax(self.attention(lstm_output), dim=1)
        # attention_weights shape: (batch_size, seq_len, 1)
        
        # Apply attention weights
        context_vector = torch.sum(attention_weights * lstm_output, dim=1)
        # context_vector shape: (batch_size, hidden_dim)
        
        return context_vector, attention_weights


class ScaledDotProductAttention(nn.Module):
    """
    Scaled Dot-Product Attention
    """
    def __init__(self, hidden_dim):
        super(ScaledDotProductAttention, self).__init__()
        self.hidden_dim = hidden_dim
        self.scale = np.sqrt(hidden_dim)
        
    def forward(self, query, key, value):
        # query, key, value shape: (batch_size, seq_len, hidden_dim)
        
        # Calculate attention scores
        scores = torch.matmul(query, key.transpose(-2, -1)) / self.scale
        # scores shape: (batch_size, seq_len, seq_len)
        
        # Apply softmax
        attention_weights = torch.softmax(scores, dim=-1)
        
        # Apply attention to values
        context_vector = torch.matmul(attention_weights, value)
        # context_vector shape: (batch_size, seq_len, hidden_dim)
        
        return context_vector, attention_weights


class MultiHeadAttention(nn.Module):
    """
    Multi-Head Attention mechanism
    """
    def __init__(self, hidden_dim, num_heads=4):
        super(MultiHeadAttention, self).__init__()
        assert hidden_dim % num_heads == 0, "hidden_dim must be divisible by num_heads"
        
        self.hidden_dim = hidden_dim
        self.num_heads = num_heads
        self.head_dim = hidden_dim // num_heads
        
        self.query_linear = nn.Linear(hidden_dim, hidden_dim)
        self.key_linear = nn.Linear(hidden_dim, hidden_dim)
        self.value_linear = nn.Linear(hidden_dim, hidden_dim)
        self.output_linear = nn.Linear(hidden_dim, hidden_dim)
        
        self.scale = np.sqrt(self.head_dim)
        
    def forward(self, x):
        batch_size, seq_len, _ = x.size()
        
        # Linear projections
        Q = self.query_linear(x)
        K = self.key_linear(x)
        V = self.value_linear(x)
        
        # Reshape for multi-head attention
        Q = Q.view(batch_size, seq_len, self.num_heads, self. head_dim).transpose(1, 2)
        K = K.view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        V = V.view(batch_size, seq_len, self. num_heads, self.head_dim).transpose(1, 2)
        
        # Scaled dot-product attention
        scores = torch.matmul(Q, K.transpose(-2, -1)) / self.scale
        attention_weights = torch.softmax(scores, dim=-1)
        attention_output = torch.matmul(attention_weights, V)
        
        # Reshape and apply output linear
        attention_output = attention_output.transpose(1, 2).contiguous().view(batch_size, seq_len, self.hidden_dim)
        output = self.output_linear(attention_output)
        
        return output, attention_weights

print("✓ Attention mechanisms defined successfully!")
print("  - AttentionLayer (Simple)")
print("  - ScaledDotProductAttention")
print("  - MultiHeadAttention")
```

## LSTM + Attention Model

```python
# LSTM with Attention mechanism
print("="*80)
print("DEFINING LSTM + ATTENTION MODEL")
print("="*80)

class LSTMAttention(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim, dropout=0.3, attention_type='simple'):
        super(LSTMAttention, self).__init__()
        
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.attention_type = attention_type
        
        # Input projection
        self.input_projection = nn.Linear(input_dim, hidden_dim)
        self.bn_input = nn.BatchNorm1d(hidden_dim)
        
        # LSTM layers
        self.lstm = nn. LSTM(
            hidden_dim, 
            hidden_dim, 
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0,
            bidirectional=False
        )
        
        # Attention mechanism
        if attention_type == 'simple':
            self.attention = AttentionLayer(hidden_dim)
        elif attention_type == 'scaled':
            self.attention = ScaledDotProductAttention(hidden_dim)
        elif attention_type == 'multihead':
            self.attention = MultiHeadAttention(hidden_dim, num_heads=4)
        
        # Fully connected layers
        self.fc1 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.bn1 = nn.BatchNorm1d(hidden_dim // 2)
        self.dropout1 = nn.Dropout(dropout)
        
        self. fc2 = nn.Linear(hidden_dim // 2, hidden_dim // 4)
        self.bn2 = nn.BatchNorm1d(hidden_dim // 4)
        self.dropout2 = nn.Dropout(dropout)
        
        self.fc3 = nn.Linear(hidden_dim // 4, output_dim)
        
        self.relu = nn.ReLU()
        
    def forward(self, x):
        batch_size = x.size(0)
        
        # Project input
        x = self.input_projection(x)
        x = self.bn_input(x)
        x = self.relu(x)
        
        # Add sequence dimension (create temporal sequences)
        x = x.unsqueeze(1). repeat(1, 10, 1)  # seq_len = 10
        
        # LSTM
        lstm_out, (hidden, cell) = self.lstm(x)
        # lstm_out shape: (batch_size, seq_len, hidden_dim)
        
        # Apply attention
        if self.attention_type == 'simple':
            context_vector, attention_weights = self. attention(lstm_out)
        elif self.attention_type == 'scaled':
            context_vector, attention_weights = self.attention(lstm_out, lstm_out, lstm_out)
            context_vector = context_vector.mean(dim=1)
        elif self.attention_type == 'multihead':
            context_vector, attention_weights = self.attention(lstm_out)
            context_vector = context_vector. mean(dim=1)
        
        # Fully connected layers
        x = self. relu(self.bn1(self.fc1(context_vector)))
        x = self.dropout1(x)
        
        x = self.relu(self.bn2(self.fc2(x)))
        x = self.dropout2(x)
        
        x = self.fc3(x)
        
        return x, attention_weights

# Model parameters
input_dim = X_train_balanced.shape[1]
output_dim = len(np.unique(y_train))
hidden_dim = 256
num_layers = 2
dropout = 0.3

print(f"\nModel Configuration:")
print(f"  Input dimension: {input_dim}")
print(f"  Output dimension: {output_dim}")
print(f"  Hidden dimension: {hidden_dim}")
print(f"  Number of LSTM layers: {num_layers}")
print(f"  Dropout: {dropout}")
print(f"  Attention type: Simple")

# Create model
lstm_attention_model = LSTMAttention(
    input_dim, hidden_dim, num_layers, output_dim, dropout, attention_type='simple'
). to(device)

print(f"\n✓ LSTM + Attention Model Created")

# Count parameters
total_params = sum(p.numel() for p in lstm_attention_model.parameters())
trainable_params = sum(p. numel() for p in lstm_attention_model.parameters() if p.requires_grad)
print(f"\nModel Statistics:")
print(f"  Total parameters: {total_params:,}")
print(f"  Trainable parameters: {trainable_params:,}")
print(f"  Model size: {total_params * 4 / 1024**2:.2f} MB")
```

```python
# Training setup for LSTM + Attention
print("="*80)
print("TRAINING: LSTM + ATTENTION MODEL")
print("="*80)

criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(lstm_attention_model.parameters(), lr=0.001, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=5)

# Prepare data loaders
batch_size = 512  # Smaller batch for better generalization

# Convert to tensors - KEEP ON CPU FIRST
X_train_tensor = torch.FloatTensor(X_train_minmax_balanced)
y_train_tensor = torch.LongTensor(y_train_balanced)
X_val_tensor = torch.FloatTensor(X_val_minmax)
y_val_tensor = torch.LongTensor(y_val. values)
X_test_tensor = torch.FloatTensor(X_test_minmax)
y_test_tensor = torch.LongTensor(y_test. values)

print(f"\nDataset sizes:")
print(f"  Train: {X_train_tensor.shape}")
print(f"  Val: {X_val_tensor.shape}")
print(f"  Test: {X_test_tensor. shape}")

# Create datasets
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)

# DataLoader WITHOUT pin_memory (since we're handling GPU transfer manually)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0, pin_memory=False)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=0, pin_memory=False)

print(f"  Batches per epoch (train): {len(train_loader)}")
print(f"  Batches per epoch (val): {len(val_loader)}")

# Training function
def train_epoch_attention(model, loader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    
    for X_batch, y_batch in loader:
        # Move to device
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)
        
        optimizer.zero_grad()
        outputs, _ = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss. backward()
        
        # Gradient clipping
        torch.nn. utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        optimizer.step()
        
        total_loss += loss.item()
        _, predicted = torch.max(outputs. data, 1)
        total += y_batch.size(0)
        correct += (predicted == y_batch).sum().item()
    
    return total_loss / len(loader), 100 * correct / total

# Validation function
def validate_attention(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for X_batch, y_batch in loader:
            # Move to device
            X_batch = X_batch.to(device)
            y_batch = y_batch.to(device)
            
            outputs, _ = model(X_batch)
            loss = criterion(outputs, y_batch)
            
            total_loss += loss.item()
            _, predicted = torch. max(outputs.data, 1)
            total += y_batch.size(0)
            correct += (predicted == y_batch). sum().item()
    
    return total_loss / len(loader), 100 * correct / total

# Training loop
num_epochs = 100
best_val_acc = 0
best_val_f1 = 0
patience = 15
patience_counter = 0
train_losses_lstm, val_losses_lstm = [], []
train_accs_lstm, val_accs_lstm = [], []

print("\n" + "="*80)
print("Starting Training...")
print("="*80)
start_time = time.time()

for epoch in range(num_epochs):
    train_loss, train_acc = train_epoch_attention(lstm_attention_model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = validate_attention(lstm_attention_model, val_loader, criterion, device)
    
    train_losses_lstm.append(train_loss)
    val_losses_lstm.append(val_loss)
    train_accs_lstm.append(train_acc)
    val_accs_lstm.append(val_acc)
    
    scheduler.step(val_acc)
    
    if (epoch + 1) % 5 == 0 or epoch == 0:
        print(f"Epoch [{epoch+1:3d}/{num_epochs}] - "
              f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}% | "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")
    
    # Early stopping with best model saving
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        patience_counter = 0
        torch.save(lstm_attention_model.state_dict(), 'models/lstm_attention_best.pth')
        if (epoch + 1) % 5 != 0:
            print(f"  → Best model saved at epoch {epoch+1}!  (Val Acc: {best_val_acc:.2f}%)")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"\n⚠ Early stopping triggered at epoch {epoch+1}")
            break

training_time_lstm = time.time() - start_time
print(f"\n{'='*80}")
print(f"Training completed in {training_time_lstm/60:.2f} minutes ({training_time_lstm:.2f} seconds)")
print(f"Best validation accuracy: {best_val_acc:.2f}%")
print(f"{'='*80}")

# Load best model
lstm_attention_model.load_state_dict(torch.load('models/lstm_attention_best.pth'))

# Clear cache
if torch.cuda.is_available():
    torch.cuda.empty_cache()
gc.collect()

print("\n✓ LSTM + Attention training completed!")
```

## GRU + Attention Model

```python
# GRU with Attention mechanism
print("="*80)
print("DEFINING GRU + ATTENTION MODEL")
print("="*80)

class GRUAttention(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim, dropout=0.3, attention_type='simple'):
        super(GRUAttention, self).__init__()
        
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.attention_type = attention_type
        
        # Input projection
        self.input_projection = nn.Linear(input_dim, hidden_dim)
        self.bn_input = nn.BatchNorm1d(hidden_dim)
        
        # GRU layers
        self.gru = nn.GRU(
            hidden_dim, 
            hidden_dim, 
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0,
            bidirectional=False
        )
        
        # Attention mechanism
        if attention_type == 'simple':
            self.attention = AttentionLayer(hidden_dim)
        elif attention_type == 'scaled':
            self.attention = ScaledDotProductAttention(hidden_dim)
        elif attention_type == 'multihead':
            self.attention = MultiHeadAttention(hidden_dim, num_heads=4)
        
        # Fully connected layers
        self.fc1 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.bn1 = nn.BatchNorm1d(hidden_dim // 2)
        self.dropout1 = nn.Dropout(dropout)
        
        self. fc2 = nn.Linear(hidden_dim // 2, hidden_dim // 4)
        self.bn2 = nn.BatchNorm1d(hidden_dim // 4)
        self.dropout2 = nn.Dropout(dropout)
        
        self.fc3 = nn.Linear(hidden_dim // 4, output_dim)
        
        self.relu = nn.ReLU()
        
    def forward(self, x):
        batch_size = x.size(0)
        
        # Project input
        x = self.input_projection(x)
        x = self.bn_input(x)
        x = self.relu(x)
        
        # Add sequence dimension
        x = x.unsqueeze(1).repeat(1, 10, 1)  # seq_len = 10
        
        # GRU
        gru_out, hidden = self.gru(x)
        # gru_out shape: (batch_size, seq_len, hidden_dim)
        
        # Apply attention
        if self.attention_type == 'simple':
            context_vector, attention_weights = self.attention(gru_out)
        elif self.attention_type == 'scaled':
            context_vector, attention_weights = self.attention(gru_out, gru_out, gru_out)
            context_vector = context_vector.mean(dim=1)
        elif self.attention_type == 'multihead':
            context_vector, attention_weights = self.attention(gru_out)
            context_vector = context_vector. mean(dim=1)
        
        # Fully connected layers
        x = self. relu(self.bn1(self.fc1(context_vector)))
        x = self.dropout1(x)
        
        x = self.relu(self.bn2(self.fc2(x)))
        x = self.dropout2(x)
        
        x = self.fc3(x)
        
        return x, attention_weights

# Create model
gru_attention_model = GRUAttention(
    input_dim, hidden_dim, num_layers, output_dim, dropout, attention_type='simple'
). to(device)

print(f"\n✓ GRU + Attention Model Created")

# Count parameters
total_params_gru = sum(p.numel() for p in gru_attention_model.parameters())
trainable_params_gru = sum(p.numel() for p in gru_attention_model.parameters() if p.requires_grad)

print(f"\nModel Statistics:")
print(f"  Total parameters: {total_params_gru:,}")
print(f"  Trainable parameters: {trainable_params_gru:,}")
print(f"  Model size: {total_params_gru * 4 / 1024**2:.2f} MB")

# Compare with LSTM
lstm_params = sum(p.numel() for p in lstm_attention_model.parameters())
print(f"\nParameter Comparison:")
print(f"  LSTM + Attention: {lstm_params:,}")
print(f"  GRU + Attention:  {total_params_gru:,}")
print(f"  Difference: {lstm_params - total_params_gru:,} ({((lstm_params - total_params_gru)/lstm_params)*100:.1f}% fewer in GRU)")
```

## Cell 25: ROC Curves

```python
# ROC Curves for all models with batch processing
print("="*80)
print("ROC CURVES")
print("="*80)

from sklearn.preprocessing import label_binarize

n_classes = len(le_target.classes_)

# Get probability predictions
model_probas = {}

# XGBoost
print("Getting XGBoost probabilities...")
model_probas['XGBoost'] = xgb_baseline.predict_proba(X_test_scaled)

# Random Forest
print("Getting Random Forest probabilities...")
model_probas['Random Forest'] = rf_baseline.predict_proba(X_test_scaled)

# Logistic Regression
print("Getting Logistic Regression probabilities...")
model_probas['Logistic Regression'] = lr_baseline.predict_proba(X_test_scaled)

# LSTM + Attention (batch processing)
print("Getting LSTM + Attention probabilities...")
lstm_attention_model. eval()
lstm_probas = []
with torch.no_grad():
    for X_batch, _ in test_loader:
        X_batch = X_batch.to(device)
        outputs = lstm_attention_model(X_batch)[0]
        probas = torch.softmax(outputs, dim=1). cpu().numpy()
        lstm_probas.append(probas)
model_probas['LSTM + Attention'] = np.vstack(lstm_probas)

# GRU + Attention (batch processing)
print("Getting GRU + Attention probabilities...")
gru_attention_model. eval()
gru_probas = []
with torch.no_grad():
    for X_batch, _ in test_loader:
        X_batch = X_batch.to(device)
        outputs = gru_attention_model(X_batch)[0]
        probas = torch.softmax(outputs, dim=1).cpu(). numpy()
        gru_probas.append(probas)
model_probas['GRU + Attention'] = np.vstack(gru_probas)

print("✓ All probabilities collected")

# Plot ROC curves
if n_classes == 2:
    # Binary classification
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    axes = axes.ravel()
    
    for idx, (model_name, y_proba) in enumerate(model_probas.items()):
        fpr, tpr, _ = roc_curve(y_test, y_proba[:, 1])
        roc_auc = auc(fpr, tpr)
        
        axes[idx].plot(fpr, tpr, label=f'AUC = {roc_auc:.3f}', linewidth=3, color='#4ECDC4')
        axes[idx].plot([0, 1], [0, 1], 'k--', linewidth=2, label='Random')
        axes[idx].set_xlabel('False Positive Rate', fontsize=12, fontweight='bold')
        axes[idx].set_ylabel('True Positive Rate', fontsize=12, fontweight='bold')
        axes[idx].set_title(f'ROC Curve - {model_name}', fontsize=13, fontweight='bold')
        axes[idx].legend(loc='lower right', fontsize=11)
        axes[idx].grid(True, alpha=0.3, linestyle='--')
    
    axes[5].axis('off')
    
else:
    # Multi-class - plot for best model only
    fig, ax = plt.subplots(figsize=(10, 8))
    
    best_model_name = results_df.loc[results_df['Test_F1']. idxmax(), 'Model']
    y_proba = model_probas[best_model_name]
    
    # Binarize the output
    y_test_bin = label_binarize(y_test, classes=range(n_classes))
    
    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']
    
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        
        ax.plot(fpr[i], tpr[i], linewidth=2, label=f'Class {i} (AUC = {roc_auc[i]:.3f})',
               color=colors[i % len(colors)])
    
    ax.plot([0, 1], [0, 1], 'k--', linewidth=2, label='Random')
    ax.set_xlabel('False Positive Rate', fontsize=12, fontweight='bold')
    ax.set_ylabel('True Positive Rate', fontsize=12, fontweight='bold')
    ax.set_title(f'ROC Curves - {best_model_name}', fontsize=14, fontweight='bold')
    ax.legend(loc='lower right', fontsize=10)
    ax.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.savefig('plots/09_roc_curves.png', dpi=300, bbox_inches='tight')
plt. show()

# Clean up
if torch.cuda.is_available():
    torch.cuda.empty_cache()
gc.collect()

print("\n✓ ROC curves generated!")
```
